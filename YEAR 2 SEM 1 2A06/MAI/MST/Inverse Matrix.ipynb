```python
import numpy as np
np.set_printoptions(precision=3, suppress=True)

# Matrix Addition
print('\nMatrix Addition')
A = np.array([[5,8],[6,9]])
print(A)
B = np.array([[3,2],[4,-1]])
print(B)
C = A+B
print(C)

# Matrix Subtraction
print('\nMatrix Subtraction')
D = A - B
print(D)

# Matrix-Scalar Multiplication
print('\nMatrix-Scalar Multiplication')
A = np.array([[5,8],[6,9]])
print(A)
b = 2
E = A*2
print(E)

# Matrix-Vector Multiplication
print('\nMatrix Multiplication')
A = np.array([[5,8],[6,9]])
print(A)
v = np.array([[3,2],[4,-1]])
print(v)
print(A@v)

# Taking transpose
print('\nTaking transpose')
print(A)
B = A.T
print(B)

# Matrix Inverse
print('\nMatrix Inverse')
print(A)
B = np.linalg.inv(A)
print(B)

# Identity Matrix
print('\nIdentity Matrix')
I3 = np.eye(3)
print(I3)
```
